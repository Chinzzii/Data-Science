```python
list = ['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, "no": 0})

housing[list] = housing[list].apply(binary_map)
```

```python
housing.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'])
status.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'], drop_first = True)
housing = pd.concat([housing, status], axis = 1)
housing.drop(['furnishingstatus'], axis = 1, inplace = True)
housing.head()
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(housing, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_train[num] = scaler.fit_transform(df_train[num])
df_train.head()
```
