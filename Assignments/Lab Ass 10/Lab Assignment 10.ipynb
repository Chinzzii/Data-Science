```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head(10)
```

```python
df.Species.value_counts()
```

```python
df.isnull().sum()
```

```python
df['Id'] = df.index+100
df.head()
```

```python
df.columns
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for i in enumerate(features):
    print(i)
```

```python
plt.figure(figsize = (10, 10))
for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    sns.histplot(data=df, x=df[i[1]], kde=True, element="step")
```

```python
plt.figure(figsize = (15, 15))
for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    sns.histplot(data=df, x=df[i[1]], hue='Species', kde=True, element="step")
```

```python
plt.figure(figsize = (16, 8))
for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    sns.boxplot(data=df, x=df[i[1]])
```
