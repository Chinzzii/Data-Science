```python
import pandas as pd
titanic = pd.read_csv("train.csv")
```

```python
titanic.head(10)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(12,10)})
import matplotlib.pyplot as plt
```

```python
sns.catplot(x="Sex", hue="Survived", kind="count", data=titanic)
```

```python
sns.catplot(x="Sex", y="Age", hue="Survived", data=titanic, dodge="true", jitter=True)
```

```python
sns.violinplot(x="Sex", y="Age", hue="Survived", data=titanic, split=True)
```

```python
sns.catplot(x="Embarked", hue="Survived", kind="count", data=titanic)
```

```python
sns.catplot(x="Boarded", hue="Survived", kind="count", data=titanic)
```

```python
sns.countplot(x="Pclass", hue="Survived", data=titanic)
```

```python
sns.swarmplot(x='Pclass',y='Age', hue="Survived", data=titanic, split=True)
```

```python
sns.boxplot(x='Pclass',y='Age', hue="Survived", data=titanic)
```

```python
sns.countplot(x="Lifeboat", data=titanic)
```

```python
sns.countplot(x="Age", hue="Survived", data=titanic)
```

```python
group = titanic.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived, annot = True, fmt ="d")
```

```python
sns.histplot(titanic["Fare"], kde=True) 
```

```python
sns.jointplot(x='Fare',y='Age',hue="Survived", data=titanic)
```

```python
g = sns.FacetGrid(data=titanic, col='Sex')
g.map(sns.displot, 'Age', kde=True)
```
